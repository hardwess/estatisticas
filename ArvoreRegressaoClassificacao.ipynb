```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

import sklearn.tree as tr
import matplotlib.pyplot as pyp
```

```python
X, y, target, _, features, _, _, _ = load_iris().values()

X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.3, random_state=6)

model1 = DecisionTreeClassifier(max_depth=2, random_state=8)
model2 = DecisionTreeClassifier(max_depth=4, random_state=0)

model1.fit(X_treino, y_treino)
model2.fit(X_treino, y_treino)
```

```python
Fig, axes = pyp.subplots(2, 1, figsize=(16, 8))
arv1 = tr.plot_tree(model1, filled=True, ax=axes[0])
arv2 = tr.plot_tree(model2, filled=True, ax=axes[1])
y_prev1 = model1.predict(X_teste)
y_prev2 = model2.predict(X_teste)
axes[0].set.title('Acurácia: [0:2.of]%'.format(accuracy_score(y_teste, y_prev1)*100))
axes[1].set.title("Acurácia: [0:2.of]%".format(accuracy_score(y_teste, y_prev2)*100))
```
