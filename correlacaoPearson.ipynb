```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
x = np.arange(20, 30)
y = np.array([3, 1, 4, 5, 14, 12, 22, 45, 89, 62])
r = np.corrcoef(x, y)

z = np.column_stack((x, y)) # Uma matriz de dados composta pelos valores das variaveis x e y

dataset = pd.DataFrame(z, columns=["x", "y"]) # Um dataframe pandas é criado com os valores da matriz z

#sns.lmplot(x="x", y="y", data=dataset) # Gráfico de dispersão
sns.heatmap(dataset.corr(), vmin = -1, vmax = 1) # Mapa de calor
```
