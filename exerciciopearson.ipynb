```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
x = np.array([0,1,3,5,8])
y = np.array([2,4,6,5,9])

r = np.corrcoef(x, y)

z = np.column_stack((x, y))

dataset = pd.DataFrame(z, columns=["x", "y"])

sns.heatmap(dataset.corr(), vmin = -1, vmax = 1)
```
